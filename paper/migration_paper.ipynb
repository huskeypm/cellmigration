```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt

class empty:pass
```

```python
def ProcessGroup(df,key):
  sdf = df.loc[(df['tag'] == key)]

  # find unique conf Val
  uniqueVals = np.unique( sdf['condVal'] )

  Ds = np.zeros_like(uniqueVals)
  Dstds = np.zeros_like(uniqueVals)
  for i,uniqueVal in enumerate( uniqueVals ) :
    ssdf = sdf.loc[sdf['condVal']==uniqueVal ]
    Dvals = ssdf['D']
    Di = np.mean(Dvals)
    Dstdi = np.std(Dvals)
    #print(key,uniqueVal,np.mean(Dvals))
    Ds[i]=Di
    Dstds[i]=Dstdi

  return uniqueVals,Ds,Dstds
```

```python

def PlotKey(conds,Ds,Dstds,key,color='k'):
  plt.plot(conds,Ds,color+'-')

  err = Dstds
  plt.fill_between(
    conds,
    Ds -err,
    Ds +err,
      alpha=0.2,
      facecolor= color
  )
  plt.scatter(conds,Ds,facecolor=color)
  plt.xlabel(key)
  plt.ylabel("D")
  plt.gcf().savefig(key+".png")

  plt.xlabel(key)
  plt.ylabel("D")
  plt.gcf().savefig(key+".png")
```

### Case: no crowders

```python
dataFrameName = "fig4.csv"
df = pd.read_csv(dataFrameName)
cases = dict()

```

```python

# get all entries with a given key name
key = "nParticles"

case = empty()
case.conds,case.Ds,case.Dstds = ProcessGroup(df,key)
cases[key] = case

normX = case.conds/case.conds[0]
PlotKey(conds,case.Ds,case.Dstds,key)

```
